```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
r = 2.0
a, b = (0, 0)
c, d = (2, 2)
```

```python
theta = np.arange(0, 2*np.pi, 0.01)
x = a+r * np.cos(theta)
y = b+r * np.sin(theta)
m = c+r * np.cos(theta)
n = d+r * np.sin(theta)
plt.plot(x,y,"red")+plt.plot(m,n,"blue")
plt.axis('equal')
```
